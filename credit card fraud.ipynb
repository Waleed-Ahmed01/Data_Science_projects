```python
import pandas as pd
import numpy as np
data=pd.read_csv('creditcard.csv')
data.head(20)
```

```python
#the dataset contains anonymous data due to privacy reasons and the features V1-V28 are pca transformed
# there cannot be much analysis done around these features
#as the dataset shows the column 'class'.A value of 0 denotes normal transaction and 1 denotes fraudulent 
class_count=data['Class'].value_counts()
class_count.plot(kind='bar')
```

```python
fraud=data[data['Class']==1]
normal=data[data['Class']==0]
fraud.shape
normal.shape
#normal transactions are 284315 as compared to only 492 of fraud ones
```

```python
#plotting the heatmap to check the co relation of all the features with one another
import seaborn as sns
import matplotlib.pyplot as plt
correlation_matrix=data.corr()
fig=plt.figure(figsize=(12,9))
sns.heatmap(correlation_matrix,vmax=0.7,square=True)
plt.show
```

```python
#Since our classes are highly skewed we should make them equivalent in order to have a normal distribution of the classes.
# we are goin to upsample the minority class of fraud case 1 
     
from sklearn.utils import resample
X=data.drop('Class',axis=1)
Y=data['Class']
#splitting the data first so that the same data doesnot go into train and test set and provide false results
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=25)
#putting together the train data
X=pd.concat([X_train,Y_train],axis=1)
#separating the two classes
fraud=X[X['Class']==1]
normal=X[X['Class']==0]
#upsampling fraud samples to be equal to normal samples
upsampled_fraud = resample(fraud,replace=True, n_samples=len(normal), random_state=25) 

#combine the normal and the upsampled fraud sample 
upsampled=pd.concat([normal,upsampled_fraud])

upsampled['Class'].value_counts()
#now the classes are balanced
upsampled.head()
```

```python
#create X_train and y_train from the upsampled data
X_train=upsampled.loc[:,'Time':'Amount']
Y_train=upsampled['Class']
from sklearn.linear_model import LogisticRegression
Lr=LogisticRegression(solver='liblinear')
upsampled=Lr.fit(X_train,Y_train)
prediction=Lr.predict(X_test)
print(prediction)
```

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix
print('Accuracy:',accuracy_score(Y_test,prediction))
```

```python
plt.figure(figsize=(12,12))
confusion_mat=confusion_matrix(Y_test,prediction)
Labels=['Normal','Fraud']
sns.heatmap(confusion_mat,xticklabels=Labels,yticklabels=Labels,annot=True,fmt="d");
plt.title("Confusion Matrix")
plt.ylabel("true class")
plt.xlabel("predicted class")
plt.show()
```
